```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np



# Definir estilo mais moderno
plt.style.use('ggplot')
sns.set_palette("colorblind")

# Extrair a contagem de atividades por companhia a√©rea
airline_counts = df['Operating Airline'].value_counts()

# Criar um novo DataFrame com as companhias e contagens
df_airlines = pd.DataFrame({
    'Companhia A√©rea': airline_counts.index,
    'N√∫mero de Atividades': airline_counts.values
})

# Ordenar o DataFrame pela contagem em ordem decrescente
df_airlines = df_airlines.sort_values(by='N√∫mero de Atividades', ascending=False)

# Criar figura com tamanho adequado
plt.figure(figsize=(18, 10))

# GR√ÅFICO 1: Top 15 companhias a√©reas (ao inv√©s de 50 que ficava ileg√≠vel)
plt.subplot(1, 2, 1)
top_15 = df_airlines.head(15)
bars = plt.barh(top_15['Companhia A√©rea'], top_15['N√∫mero de Atividades'], 
         color=sns.color_palette("Blues_d", 15))

# Adicionar valores nas barras
for i, bar in enumerate(bars):
    plt.text(bar.get_width() + 50, bar.get_y() + bar.get_height()/2, 
             f'{top_15["N√∫mero de Atividades"].iloc[i]:,}',
             va='center', fontsize=9)

plt.title('Top 15 Companhias A√©reas por Volume de Atividades', fontsize=14, fontweight='bold')
plt.xlabel('N√∫mero de Atividades', fontsize=12)
plt.gca().invert_yaxis()  # Inverter o eixo y para mostrar a maior contagem no topo
plt.grid(axis='x', linestyle='--', alpha=0.7)

# GR√ÅFICO 2: Distribui√ß√£o por grupos de companhias
plt.subplot(1, 2, 2)

# Criar grupos mais significativos
top_10 = df_airlines.head(10)
next_10 = df_airlines.iloc[10:20]
other_airlines = df_airlines.iloc[20:]

# Criar dados agregados para o gr√°fico de pizza
sizes = [
    top_10['N√∫mero de Atividades'].sum(),
    next_10['N√∫mero de Atividades'].sum(),
    other_airlines['N√∫mero de Atividades'].sum()
]

labels = [
    f'Top 10 Companhias\n{sizes[0]:,} atividades ({sizes[0]/sum(sizes):.1%})',
    f'Companhias 11-20\n{sizes[1]:,} atividades ({sizes[1]/sum(sizes):.1%})',
    f'Outras Companhias\n{sizes[2]:,} atividades ({sizes[2]/sum(sizes):.1%})'
]

# Criar gr√°fico de pizza com espa√ßamento entre as fatias
explode = (0.05, 0.05, 0.05)
wedges, texts, autotexts = plt.pie(sizes, labels=labels, autopct='%1.1f%%', 
                                   startangle=90, explode=explode, 
                                   shadow=True, textprops={'fontsize': 11})

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')

plt.axis('equal')
plt.title('Distribui√ß√£o de Atividades por Grupo de Companhias', fontsize=14, fontweight='bold')

# Adicionar detalhes das top 5 companhias em uma caixa de texto
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
top_5_info = "\n".join([f"{row['Companhia A√©rea']}: {row['N√∫mero de Atividades']:,}" 
                        for _, row in top_15.head(5).iterrows()])
plt.text(1.4, 0, 'Top 5 Companhias:\n\n' + top_5_info, fontsize=10,
         verticalalignment='top', horizontalalignment='center',
         transform=plt.gca().transAxes, bbox=props)

# Ajustes finais
plt.tight_layout()
plt.suptitle('An√°lise de atividades de passageiros por Companhia A√©rea', fontsize=16, fontweight='bold', y=1.02)

# Adicionar informa√ß√µes √∫teis no rodap√©
plt.figtext(0.5, -0.01, 'Total de companhias analisadas: ' + str(len(df_airlines)) + 
            ' | Total de atividades: ' + f'{df_airlines["N√∫mero de Atividades"].sum():,}',
            ha='center', fontsize=10)

plt.subplots_adjust(top=0.9)
plt.show()
```

# 4. Prepara√ß√£o dos Dados

## 4.1 An√°lise Bidimensional

### 4.2 Respondendo as Perguntas de Neg√≥cio:

 **A)Quais caracter√≠sticas das aeronaves mais influenciam a opera√ß√£o em rotas internacionais?**:

 

As an√°lises bidimensionais s√£o √∫teis para identificar poss√≠veis rela√ß√µes entre duas vari√°veis. Elas ajudam a entender como diferentes vari√°veis interagem e como essas intera√ß√µes podem influenciar rotas internacionais??.


**WOE (Weight of Evidence) √© a estimativa quantitativa da probabilidade de comprometimento para linhas de evid√™ncia individuais e m√∫ltiplas.**

Para que possamos identificar quais fatores possuem uma forte rela√ß√£o com rotas internacionais?? e selecionar os dados que ser√£o preparados, vou realizar uma an√°lise bidimensional atrav√©s do Information Value (IV).

C√°lculo do Information Value

O c√°lculo do Information Value (IV), uma t√©cnica estat√≠stica utilizada para avaliar o poder preditivo de vari√°veis independentes em rela√ß√£o a uma vari√°vel de resultado.

Quanto maior o IV de uma vari√°vel, maior √© a sua capacidade de discriminar entre as categorias da vari√°vel resultado.

Os valores de IV podem ser interpretados da seguinte forma:















```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ================== FUN√á√ïES AUXILIARES ==================

def preparar_variavel_numerica(df, feature, bins=5):
    """Agrupa vari√°veis num√©ricas em faixas usando quantis."""
    if pd.api.types.is_numeric_dtype(df[feature]):
        # Criar nome da nova coluna binarizada
        new_col = f"{feature}_bin"
        df[new_col] = pd.qcut(df[feature], q=bins, duplicates='drop').astype(str)
        return new_col
    return feature

def calcular_woe_iv(df, feature, target, woe_tables):
    """Calcula Weight of Evidence (WoE) e Information Value (IV)."""
    try:
        # Criar tabela cruzada
        cross_tab = pd.crosstab(df[feature], df[target], margins=False)
        
        # Adicionar pequena constante para evitar divis√£o por zero
        cross_tab = cross_tab.replace(0, 0.5)
        
        # Calcular distribui√ß√µes
        total_0 = cross_tab[0].sum()
        total_1 = cross_tab[1].sum()
        
        cross_tab['%_Domestic'] = cross_tab[0] / total_0
        cross_tab['%_International'] = cross_tab[1] / total_1
        
        # Calcular WoE e IV
        cross_tab['WoE'] = np.log(cross_tab['%_Domestic'] / cross_tab['%_International'])
        cross_tab['IV'] = (cross_tab['%_Domestic'] - cross_tab['%_International']) * cross_tab['WoE']
        
        # Formatar para exibi√ß√£o
        cross_tab['%_Domestic'] = (cross_tab['%_Domestic'] * 100).round(1).astype(str) + '%'
        cross_tab['%_International'] = (cross_tab['%_International'] * 100).round(1).astype(str) + '%'
        
        # Armazenar resultados
        woe_tables[feature] = cross_tab.copy()
        return cross_tab['IV'].sum()
    
    except Exception as e:
        print(f"Erro ao processar {feature}: {str(e)}")
        return 0

def classificar_poder_preditivo(iv_value):
    """Classifica o poder preditivo com base no IV."""
    if iv_value < 0.02: return 'Sem poder'
    elif iv_value < 0.1: return 'Fraco'
    elif iv_value < 0.3: return 'Moderado'
    elif iv_value < 0.5: return 'Forte'
    else: return 'Muito forte'

# ================== EXECU√á√ÉO PRINCIPAL ==================

# 1. Carregar e preparar os dados
print("‚è≥ Carregando e preparando dados...")
try:
    df = pd.read_csv("Air_Traffic_Landings_Statistics.csv")
    df['Target'] = (df['GEO Summary'] == 'International').astype(int)  # 1=Internacional, 0=Dom√©stico
except Exception as e:
    print(f"Erro ao carregar dados: {e}")
    raise

# 2. Selecionar e preparar features
features = [
    'Operating Airline',
    'GEO Region',
    'Landing Aircraft Type',
    'Aircraft Body Type',
    'Aircraft Manufacturer',
    'Total Landed Weight'
]

# Dicion√°rio para mapear nomes originais para nomes processados
processed_features = {}

# Pr√©-processamento
for col in features:
    try:
        processed_name = preparar_variavel_numerica(df, col)
        processed_features[col] = processed_name
        
        # Preencher valores faltantes
        if pd.api.types.is_numeric_dtype(df[col]):
            # Tratar outliers
            q1, q3 = df[col].quantile(0.25), df[col].quantile(0.75)
            iqr = q3 - q1
            df[col] = np.clip(df[col], q1 - 1.5*iqr, q3 + 1.5*iqr)
    except Exception as e:
        print(f"Erro ao processar {col}: {e}")
        continue

# 3. Calcular WoE e IV
print("üìä Calculando WoE e IV para cada vari√°vel...")
woe_results = {}
iv_values = {}

for original_name in features:
    feature_name = processed_features.get(original_name, original_name)
    try:
        if feature_name in df.columns:
            iv = calcular_woe_iv(df, feature_name, 'Target', woe_results)
            iv_values[original_name] = iv
        else:
            print(f"Vari√°vel {feature_name} n√£o encontrada no DataFrame")
            iv_values[original_name] = 0
    except Exception as e:
        print(f"Erro ao calcular IV para {original_name}: {e}")
        iv_values[original_name] = 0

# 4. Criar tabela de resultados
iv_df = pd.DataFrame.from_dict(iv_values, orient='index', columns=['IV'])
iv_df['Poder Preditivo'] = iv_df['IV'].apply(classificar_poder_preditivo)
iv_df = iv_df.sort_values('IV', ascending=False)

# ================== VISUALIZA√á√ÉO ==================

# Gr√°fico de IV corrigido (sem warning de palette)
plt.figure(figsize=(12,6))
ax = sns.barplot(x='IV', y=iv_df.index, data=iv_df, 
                 hue=iv_df.index, palette='viridis', dodge=False, legend=False)

# Adicionar valores nas barras
for p in ax.patches:
    ax.annotate(f"{p.get_width():.3f}", 
                (p.get_width() + 0.01, p.get_y() + p.get_height()/2),
                ha='left', va='center')

plt.title('Information Value por Vari√°vel (Aloca√ß√£o de Frota)', pad=20)
plt.xlabel('Information Value (IV)')
plt.ylabel('Vari√°veis')
plt.tight_layout()
plt.show()

# ================== RESULTADOS DETALHADOS ==================

# 5. Exibir tabelas completas para vari√°veis com IV ‚â• 0.3 (Forte ou superior)
print("\nüîç RELAT√ìRIO COMPLETO PARA VARI√ÅVEIS COM PODER FORTE OU SUPERIOR (IV ‚â• 0.3)")
print("="*85)

for original_name in iv_df[iv_df['IV'] >= 0.3].index:
    feature_name = processed_features.get(original_name, original_name)
    
    if feature_name not in woe_results:
        print(f"\n‚ö†Ô∏è Dados n√£o dispon√≠veis para: {original_name}")
        continue
        
    print(f"\n‚ñ∫ Vari√°vel: {original_name} (IV = {iv_df.loc[original_name,'IV']:.3f} - {iv_df.loc[original_name,'Poder Preditivo']})")
    print("-"*60)
    
    # Preparar tabela WoE de forma segura
    try:
        woe_table = woe_results[feature_name].copy()
        cols_to_show = ['%_Domestic', '%_International', 'WoE', 'IV']
        
        # Verificar se as colunas existem
        available_cols = [col for col in cols_to_show if col in woe_table.columns]
        if not available_cols:
            print("Colunas esperadas n√£o encontradas na tabela WoE")
            continue
            
        woe_table = woe_table[available_cols]
        
        # Arredondar valores num√©ricos
        if 'WoE' in woe_table.columns:
            woe_table['WoE'] = woe_table['WoE'].round(3)
        if 'IV' in woe_table.columns:
            woe_table['IV'] = woe_table['IV'].round(5)
        
        print(woe_table.sort_values('WoE', ascending=False).to_string())
        
        # Interpreta√ß√£o
        if 'WoE' in woe_table.columns:
            top_domestic = woe_table['WoE'].idxmax()
            top_international = woe_table['WoE'].idxmin()
            print(f"\nInterpreta√ß√£o:")
            print(f"- Maior associa√ß√£o com DOM√âSTICO: {top_domestic} (WoE = {woe_table.loc[top_domestic,'WoE']:.2f})")
            print(f"- Maior associa√ß√£o com INTERNACIONAL: {top_international} (WoE = {woe_table.loc[top_international,'WoE']:.2f})")
        else:
            print("\nN√£o foi poss√≠vel gerar interpreta√ß√£o (WoE n√£o calculado)")
            
    except Exception as e:
        print(f"Erro ao processar tabela WoE: {e}")
    
    print("="*85)

```

**Premissa**

Definir baixa performance como os 25% com menos pousos.

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.metrics import classification_report, roc_auc_score
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import make_pipeline
import matplotlib.pyplot as plt
import seaborn as sns

# ================== CONFIGURA√á√ÉO ==================
QUANTIL_BAIXO = 0.25  # Definir baixa performance como os 25% com menos pousos
N_SPLITS = 5          # N√∫mero de folds para valida√ß√£o cruzada

# ================== PR√â-PROCESSAMENTO ==================
df = pd.read_csv("Air_Traffic_Landings_Statistics.csv")
df['BaixaPerformance'] = (df['Landing Count'] <= df['Landing Count'].quantile(QUANTIL_BAIXO)).astype(int)
df['Combinacao'] = df['Operating Airline'] + " + " + df['Aircraft Body Type']

# Verificar balanceamento
print("\nüîç DISTRIBUI√á√ÉO DAS CLASSES:")
print(df['BaixaPerformance'].value_counts(normalize=True))

# ================== VALIDA√á√ÉO CRUZADA ==================
X = pd.get_dummies(df[['Operating Airline', 'Aircraft Body Type', 'Total Landed Weight']])
y = df['BaixaPerformance']

    # Pipeline com SMOTE (oversampling) e RandomForest
pipeline = make_pipeline(
    SMOTE(random_state=42),  # Corrige desbalanceamento
    RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')
)

# Valida√ß√£o cruzada estratificada
cv = StratifiedKFold(n_splits=N_SPLITS, shuffle=True, random_state=42)
cv_scores = cross_val_score(pipeline, X, y, cv=cv, scoring='roc_auc')

print("\nüìä VALIDA√á√ÉO CRUZADA (AUC-ROC):")
print(f"M√©dia: {np.mean(cv_scores):.3f} | Desvio Padr√£o: {np.std(cv_scores):.3f}")
print(f"Intervalo (95%): [{np.mean(cv_scores)-1.96*np.std(cv_scores):.3f}, {np.mean(cv_scores)+1.96*np.std(cv_scores):.3f}]")

# ================== TREINO FINAL E TESTE ==================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
pipeline.fit(X_train, y_train)

# Avalia√ß√£o no conjunto de teste
y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:,1]

print("\nüìà M√âTRICAS NO CONJUNTO DE TESTE:")
print(classification_report(y_test, y_pred))
print(f"AUC-ROC: {roc_auc_score(y_test, y_prob):.3f}")

# ================== AN√ÅLISE DE SUPERAJUSTE ==================
train_prob = pipeline.predict_proba(X_train)[:,1]
print("\nüîç COMPARA√á√ÉO TREINO/TESTE:")
print(f"AUC Treino: {roc_auc_score(y_train, train_prob):.3f}")
print(f"AUC Teste: {roc_auc_score(y_test, y_prob):.3f}")
print("\nDiferen√ßa aceit√°vel (<0.05)?" , 
      "‚úÖ Sim" if abs(roc_auc_score(y_train, train_prob) - roc_auc_score(y_test, y_prob)) < 0.05 else "‚ùå N√£o")

# ================== FEATURE IMPORTANCE ==================
rf_model = pipeline.named_steps['randomforestclassifier']
importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print("\nüéØ TOP 10 VARI√ÅVEIS MAIS IMPORTANTES:")
print(importance.head(10).to_string())

# Plot
plt.figure(figsize=(12,6))
sns.barplot(x='Importance', y='Feature', data=importance.head(10), palette='viridis')
plt.title('Feature Importance (Random Forest)')
plt.tight_layout()
plt.show()

# ================== RESPOSTA √Ä PERGUNTA DE NEG√ìCIO ==================
# Calcular probabilidades para combina√ß√µes espec√≠ficas de "companhia a√©rea + tipo de aeronave"
df['Probabilidade_BaixaPerformance'] = pipeline.predict_proba(X)[:,1]

# Criar tabela com as combina√ß√µes e suas probabilidades m√©dias
combinacoes = df.groupby(['Combinacao'])['Probabilidade_BaixaPerformance'].agg(['mean', 'count']).reset_index()
combinacoes.rename(columns={'mean': 'Probabilidade_Media', 'count': 'Frequencia'}, inplace=True)

# Filtrar combina√ß√µes com maior risco (probabilidade alta)
combinacoes_risco = combinacoes[combinacoes['Probabilidade_Media'] > 0.5].sort_values(by='Probabilidade_Media', ascending=False)

print("\nüîç COMBINA√á√ïES COM MAIOR RISCO DE BAIXA PERFORMANCE:")
print(combinacoes_risco.to_string())

# Plot das combina√ß√µes com maior risco
plt.figure(figsize=(12,6))
sns.barplot(data=combinacoes_risco.head(10), x='Probabilidade_Media', y='Combinacao', palette='Reds_r')
plt.title('Combina√ß√µes com Maior Risco de Baixa Performance')
plt.xlabel('Probabilidade M√©dia de Baixa Performance')
plt.ylabel('Combina√ß√£o (Companhia A√©rea + Tipo de Aeronave)')
plt.tight_layout()
plt.show()
```

**E)Quais vari√°veis explicam melhor a escolha de aeronaves de corpo largo (wide body)?**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suprimir warnings espec√≠ficos
warnings.filterwarnings("ignore", category=FutureWarning)

# ================== FUN√á√ïES AUXILIARES ==================

def preparar_variavel_numerica(df, feature, bins=5):
    """Agrupa vari√°veis num√©ricas em faixas usando quantis."""
    if pd.api.types.is_numeric_dtype(df[feature]):
        new_col = f"{feature}_bin"
        df[new_col] = pd.qcut(df[feature], q=bins, duplicates='drop').astype(str)
        return new_col
    return feature

def calcular_woe_iv(df, feature, target, woe_tables):
    """Calcula Weight of Evidence (WoE) e Information Value (IV)."""
    try:
        cross_tab = pd.crosstab(df[feature], df[target], margins=False)
        cross_tab = cross_tab.replace(0, 0.5)
        
        total_0 = cross_tab[0].sum()
        total_1 = cross_tab[1].sum()
        cross_tab['%_NaoWideBody'] = cross_tab[0] / total_0
        cross_tab['%_WideBody'] = cross_tab[1] / total_1
        
        cross_tab['WoE'] = np.log(cross_tab['%_NaoWideBody'] / cross_tab['%_WideBody'])
        cross_tab['IV'] = (cross_tab['%_NaoWideBody'] - cross_tab['%_WideBody']) * cross_tab['WoE']
        
        cross_tab['%_NaoWideBody'] = (cross_tab['%_NaoWideBody'] * 100).round(1).astype(str) + '%'
        cross_tab['%_WideBody'] = (cross_tab['%_WideBody'] * 100).round(1).astype(str) + '%'
        
        woe_tables[feature] = cross_tab.copy()
        return cross_tab['IV'].sum()
    
    except Exception as e:
        print(f"Erro ao processar {feature}: {str(e)}")
        return 0

def classificar_poder_preditivo(iv_value):
    if iv_value < 0.02: return 'Sem poder'
    elif iv_value < 0.1: return 'Fraco'
    elif iv_value < 0.3: return 'Moderado'
    elif iv_value < 0.5: return 'Forte'
    else: return 'Muito forte'

# ================== EXECU√á√ÉO PRINCIPAL ==================

# 1. Carregar e preparar os dados
df = pd.read_csv("Air_Traffic_Landings_Statistics.csv")

# Criando nossa vari√°vel alvo: 1 para Wide Body, 0 para outros tipos
df['WideBody'] = (df['Aircraft Body Type'] == 'Wide Body').astype(int)

# 2. Selecionar features que podem explicar a escolha de aeronaves wide body
features = [
    'Operating Airline',
    'Published Airline',
    'GEO Summary',
    'GEO Region',
    'Aircraft Manufacturer',
    'Total Landed Weight',
    'Landing Count'
]

# Pr√©-processamento para vari√°veis num√©ricas
for col in features:
    if pd.api.types.is_numeric_dtype(df[col]):
        q1, q3 = df[col].quantile(0.25), df[col].quantile(0.75)
        df[col] = np.clip(df[col], q1 - 1.5*(q3-q1), q3 + 1.5*(q3-q1))

# 3. Calcular WoE e IV
woe_results = {}
iv_values = {}

for feature in features:
    feature_name = preparar_variavel_numerica(df, feature)
    iv = calcular_woe_iv(df, feature_name, 'WideBody', woe_results)
    iv_values[feature] = iv

# 4. Criar tabela de resultados
iv_df = pd.DataFrame.from_dict(iv_values, orient='index', columns=['IV'])
iv_df['Poder Preditivo'] = iv_df['IV'].apply(classificar_poder_preditivo)
iv_df = iv_df.sort_values('IV', ascending=False)

# ================== VISUALIZA√á√ÉO ==================

plt.figure(figsize=(12,6))
ax = sns.barplot(
    x='IV', 
    y=iv_df.index, 
    data=iv_df, 
    hue=iv_df.index,
    palette='viridis', 
    legend=False
)

for p in ax.patches:
    ax.annotate(
        f"{p.get_width():.3f}", 
        (p.get_width() + 0.01, p.get_y() + p.get_height()/2),
        ha='left', 
        va='center'
    )

plt.title('Fatores que Influenciam a Escolha de Aeronaves de Corpo Largo', pad=20)
plt.xlabel('Information Value (IV)')
plt.ylabel('Vari√°veis')
plt.tight_layout()
plt.show()

# ================== RESULTADOS DETALHADOS ==================

print("\nüîç TOP CATEGORIAS POR VARI√ÅVEL (IV ‚â• 0.1)")

for feature in iv_df[iv_df['IV'] >= 0.1].index:
    feature_name = feature+'_bin' if feature+'_bin' in df.columns else feature
    
    if feature_name not in woe_results:
        continue
        
    print(f"\n‚ñ∫ {feature} (IV = {iv_df.loc[feature,'IV']:.3f} - {iv_df.loc[feature,'Poder Preditivo']})")
    print("-"*60)
    
    woe_table = woe_results[feature_name].copy()
    cols = ['%_NaoWideBody', '%_WideBody', 'WoE', 'IV']
    woe_table = woe_table[[c for c in cols if c in woe_table.columns]]
    
    # Limitar a 15 registros e ordenar por WoE
    woe_table = woe_table.sort_values('WoE', ascending=True).head(15)
    
    if 'WoE' in woe_table.columns:
        woe_table['WoE'] = woe_table['WoE'].round(3)
    if 'IV' in woe_table.columns:
        woe_table['IV'] = woe_table['IV'].round(5)
    
    print(woe_table.to_string())
    
    if 'WoE' in woe_table.columns:
        print(f"\nInterpreta√ß√£o:")
        print(f"- Categoria com MAIOR tend√™ncia para Wide Body: {woe_table['WoE'].idxmin()}")
        print(f"- Categoria com MENOR tend√™ncia para Wide Body: {woe_table['WoE'].idxmax()}")
    
    print("="*85)

# Resumo executivo
print("\n‚≠ê RESUMO DAS VARI√ÅVEIS MAIS RELEVANTES PARA ESCOLHA DE AERONAVES WIDE BODY")
print(iv_df.to_string())

# Contagem total de aeronaves Wide Body vs demais tipos
total_wide = df['WideBody'].sum()
total = len(df)
percent_wide = (total_wide/total) * 100

print(f"\nEstat√≠sticas gerais:")
print(f"- Total de pousos analisados: {total}")
print(f"- Pousos com aeronaves Wide Body: {total_wide} ({percent_wide:.1f}%)")
print(f"- Pousos com outros tipos de aeronaves: {total - total_wide} ({100-percent_wide:.1f}%)")
```

```python
def gerar_pdf_scrum():
    # Nome do arquivo PDF
    pdf_filename = "Scrum_Projeto_Analise_Aeronaves.pdf"
    pdf_pages = PdfPages(pdf_filename)

    # Fun√ß√£o auxiliar para criar tabelas
    def criar_tabela(ax, data, titulo):
        ax.axis('off')
        ax.set_title(titulo, fontsize=14, fontweight='bold', pad=10)

        df = pd.DataFrame(data)
        table = ax.table(
            cellText=df.values,
            colLabels=list(df.columns),
            cellLoc='left',
            loc='center',
            colWidths=[0.2, 0.8]
        )

        table.auto_set_font_size(False)
        table.set_fontsize(10)
        table.scale(1.1, 1.5)

        for (i, j), cell in table.get_celld().items():
            cell.set_edgecolor('#3a7ca5')
            if i == 0:
                cell.set_facecolor('#2f6690')
                cell.set_text_props(color='white', weight='bold')
            else:
                cell.set_height(0.2)

    # Criar figura com 4 se√ß√µes
    fig, axs = plt.subplots(4, 1, figsize=(14, 18))

    # 1. Vis√£o do Produto
    data_visao = {
        "Item": ["Objetivo", "Valor Esperado"],
        "Descri√ß√£o": [
            "Realizar an√°lise de dados de opera√ß√µes de aeronaves para identificar padr√µes e oportunidades",
            "Fornecer insights estrat√©gicos para otimiza√ß√£o de frota e parcerias com fabricantes"
        ]
    }
    criar_tabela(axs[0], data_visao, "Vis√£o do Produto")

    # 2. Backlog do Produto (Sprints)
    data_sprints = {
        "Sprint": ["Sprint 1 (Semana 1)", 
                  "Sprint 2 (Semana 2)", 
                  "Sprint 3 (Semana 3)"],
        "Tarefas": [
            "Importa√ß√£o e limpeza dos dados\nAn√°lise descritiva inicial\nEstrutura√ß√£o dos KPIs",
            "An√°lise estat√≠stica para responder perguntas de neg√≥cio\nDesenvolvimento de modelos\nGera√ß√£o de visualiza√ß√µes",
            "Constru√ß√£o de dashboard/apresenta√ß√£o\nRecomenda√ß√µes de a√ß√µes estrat√©gicas\nPrepara√ß√£o de relat√≥rios finais"
        ]
    }
    criar_tabela(axs[1], data_sprints, "Backlog do Produto - Sprints")

    # 3. Equipe Scrum
    data_equipe = {
        "Fun√ß√£o": ["Product Owner", 
                  "Scrum Master", 
                  "Time de Desenvolvimento"],
        "Responsabilidades": [
            "Definir prioridades e garantir que a an√°lise atenda √†s necessidades do neg√≥cio",
            "Facilitar o processo √°gil e garantir o cumprimento do cronograma",
            "Analistas de dados, estat√≠sticos e especialistas em avia√ß√£o"
        ]
    }
    criar_tabela(axs[2], data_equipe, "Equipe Scrum")

    # 4. M√©tricas de Sucesso
    data_metricas = {
        "M√©trica": ["Qualidade dos dados", 
                   "Relev√¢ncia dos insights", 
                   "Clareza das visualiza√ß√µes", 
                   "Aplicabilidade das recomenda√ß√µes"],
        "Meta": [
            "Dados completos e limpos",
            "Responder a todas as perguntas de neg√≥cio",
            "Visualiza√ß√µes compreens√≠veis para decisores",
            "Recomenda√ß√µes acion√°veis pela equipe de opera√ß√µes"
        ]
    }
    criar_tabela(axs[3], data_metricas, "M√©tricas de Sucesso")

    # Ajustar layout e salvar PDF
    plt.tight_layout(pad=3.0)
    pdf_pages.savefig(fig, bbox_inches='tight')
    pdf_pages.close()

    print(f"PDF '{pdf_filename}' gerado com sucesso!")
    return pdf_filename

def visualizar_pdf(pdf_path):
    webbrowser.open(pdf_path)

if __name__ == "__main__":
    pdf_file = gerar_pdf_scrum()
    visualizar_pdf(pdf_file)
```

```python
from IPython.display import display, HTML
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import textwrap
import matplotlib.patches as patches
import os
import base64
import fitz  # PyMuPDF

# Configura√ß√µes globais de estilo
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

def converter_pdf_para_imagens(pdf_path):
    """Converte cada p√°gina do PDF para imagem PNG"""
    doc = fitz.open(pdf_path)
    imagens = []
    
    for page in doc:
        pix = page.get_pixmap(dpi=150)
        img_bytes = pix.tobytes("png")
        img_base64 = base64.b64encode(img_bytes).decode('utf-8')
        imagens.append(img_base64)
    
    return imagens

def visualizar_pdf_no_jupyter(pdf_path):
    # Converter cada p√°gina do PDF para imagem
    imagens = converter_pdf_para_imagens(pdf_path)
    
    # HTML para exibir todas as p√°ginas como imagens, uma ap√≥s a outra
    html_content = """
    <div style="display: flex; flex-direction: column; align-items: center; gap: 30px;">
    """
    
    for img_base64 in imagens:
        html_content += f"""
        <div style="width: 100%; display: flex; justify-content: center;">
            <img src="data:image/png;base64,{img_base64}" style="max-width: 100%; height: auto; border: 1px solid #ddd; box-shadow: 0 2px 5px rgba(0,0,0,0.1);">
        </div>
        """
    
    html_content += f"""
    </div>
    
    <div style="margin: 20px 0; padding: 15px; background: #f5f8fa; border-radius: 5px; text-align: center;">
        <strong style="color: #1e3d59;">Op√ß√µes:</strong>
        <a href="{pdf_path}" download style="margin-left: 15px; text-decoration: none; color: #2f6690; font-weight: bold; padding: 5px 10px; border: 1px solid #2f6690; border-radius: 3px;">‚¨áÔ∏è Baixar Relat√≥rio Completo</a>
        <a href="{pdf_path}" target="_blank" style="margin-left: 15px; text-decoration: none; color: #2f6690; font-weight: bold; padding: 5px 10px; border: 1px solid #2f6690; border-radius: 3px;">üîó Abrir em Nova Aba</a>
    </div>
    """
    
    display(HTML(html_content))

if __name__ == "__main__":
    # Gera o relat√≥rio completo (fun√ß√µes anteriores permanecem iguais)
    pdf_path = gerar_relatorio_completo()
    
    # Mostra no Jupyter
    try:
        from IPython import get_ipython
        if get_ipython() is not None:
            visualizar_pdf_no_jupyter(pdf_path)
    except:
        print(f"Relat√≥rio gerado com sucesso: {os.path.abspath(pdf_path)}")
```
